```python
# cargamos el dataframe creado:
df_perros = pd.read_excel("C:/Users/isaw9/OneDrive/Escritorio/perros.xlsx", index_col = None)
df_perros
```

```python
# Limpiar el texto "kg", eliminar espacios adicionales y convertir a tipo numérico
df_perros['Peso_medio'] = df_perros['Peso_medio'].str.replace('kg', '', regex=False)  # Eliminar "kg"
df_perros['Peso_medio'] = df_perros['Peso_medio'].str.strip()  # Eliminar espacios en blanco

# Convertir a tipo numérico (float) - se gestionan los errores automáticamente
df_perros['Peso_medio'] = pd.to_numeric(df_perros['Peso_medio'], errors='coerce')

# Mostrar el DataFrame limpio
print(df_perros)
```
