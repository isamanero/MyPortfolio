```python
# Importar librer칤as para tratamiento de datos
# -----------------------------------------------------------------------
import pandas as pd  # Pandas es una poderosa librer칤a para manipulaci칩n y an치lisis de datos en Python.

# Importar librer칤as manipulaci칩n de datos

import requests

# Importar librer칤as para automatizaci칩n de navegadores web con Selenium
# -----------------------------------------------------------------------
from selenium import webdriver  # Selenium es una herramienta para automatizar la interacci칩n con navegadores web.
from webdriver_manager.chrome import ChromeDriverManager  # ChromeDriverManager gestiona la instalaci칩n del controlador de Chrome.
from selenium.webdriver.common.keys import Keys  # Keys es 칰til para simular eventos de teclado en Selenium.
from selenium.webdriver.support.ui import Select  # Select se utiliza para interactuar con elementos <select> en p치ginas web.

# Importar librer칤as para pausar la ejecuci칩n
# -----------------------------------------------------------------------
from time import sleep  # Sleep se utiliza para pausar la ejecuci칩n del programa por un n칰mero de segundos.

# Configuraciones
# -----------------------------------------------------------------------
pd.set_option('display.max_columns', None)  # Establece una opci칩n de Pandas para mostrar todas las columnas de un DataFrame.
```

```python

# Iniciar el navegador
driver = webdriver.Chrome()
driver.maximize_window()

# Acceder a la web
driver.get("https://www.fishipedia.es/pez/type/requin")
print("Accede a la web")

lista_tiburones = [] # Aqu칤 creamos una lista vac칤a, donde se ir치n adjuntando los datos

# Hacemos un scrapeo para 32 filas:

for i in range(1,32):
    # Hacer clic en cada tibur칩n
    driver.find_element("css selector", f"#listItems > li:nth-child({i}) > div > a").click()
    print(f"Accediendo al tibur칩n {i}")


    sleep(3)  # Espera para asegurar que la p치gina haya cargado correctamente

    try:
        # Obtener el nombre cient칤fico
        nombre_cientifico = driver.find_element("css selector", "#Js--BannerJs table tbody tr.g-6-col td em").text
        print(f"Nombre cient칤fico: {nombre_cientifico}")
        familia=driver.find_element("css selector", "#Js--BannerJs > table > tbody > tr:nth-child(5) > td > a").text
        print(f"Familia: {familia}")
        genero=driver.find_element("css selector", "#Js--BannerJs > table > tbody > tr:nth-child(6) > td").text
        print(f"Genero: {genero}")
        descriptor =driver.find_element("css selector", "#Js--BannerJs > table > tbody > tr:nth-child(2) > td").text
        print(f"Descriptor: {descriptor}")
        anio =driver.find_element("css selector", "#Js--BannerJs > table > tbody > tr:nth-child(3) > td").text
        print(f"A침o: {anio}")
        IUCN_status =driver.find_element("css selector", "#Js--BannerJs > table > tbody > tr:nth-child(4) > td").text
        print(f"Status_IUCN: {IUCN_status}")
        try:
            tamano_medio=driver.find_element("css selector", "#sheetContent > section:nth-child(3) > div:nth-child(3) > div.isHiddenBeforeExLarge > ul > li.g-6-col.g-d-3-col.g-l-12-col.g-cell.inColumnDesk.ListHighlightInfo--Icon.ListHighlightInfo--Icon-size > div.ListHighlightInfo--Value.isCapitalize").text
            print(f"Tama침o medio: {tamano_medio}")
        except:
            print("sin dato")
        try:
            social=driver.find_element("css selector", "#sheetContent > section:nth-child(3) > div:nth-child(7) > ul > li.g-12-col.g-t-6-col.g-l-3-col.g-cell.ListHighlightInfo--Icon.ListHighlightInfo--Icon-sociability > div.ListHighlightInfo--Value.isCapitalize").text
            print(f"Sociabilidad: {social}")
        except:
            print("sin dato")
      
        # Guardar en la lista
        lista_tiburones.append({"Nombre_cientifico": nombre_cientifico, "Familia": familia,
                                "Genero":genero ,"Status_IUCN":IUCN_status,
                                "Tama침o_medio" :tamano_medio, "Descriptor":descriptor, "A침o": anio,
                                "Sociabilidad":social})
        
    except Exception as e:
        print(f"Error al obtener el dato {i}: {e}")

    sleep(2)  # Peque침a pausa antes de regresar
    driver.back()
    sleep(3)  # Espera a que la lista de tiburones se vuelva a cargar

# Cerrar el navegador
driver.quit()

# Imprimir resultados
print("Lista de tiburones:", lista_tiburones)
```
