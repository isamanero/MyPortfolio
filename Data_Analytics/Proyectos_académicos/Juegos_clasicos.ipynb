## Juego del ahorcado

```python
import random

conjunto_palabras_secretas = ["adalab", "animo", "python", "crack", "suerte"]
palabra_secreta = random.choice(conjunto_palabras_secretas)

lista_palabra_secreta = list(palabra_secreta) # Aqui convertimos en lista nuestra palabra secreta
lista_tablero = ["*"] * len(palabra_secreta) # Creamos el tablero, según la longitud de la palabra secreta
print(f"ESTE ES TU TABLERO DE JUEGO: 🕹️ {lista_tablero}")

intentos = 0 #Establecemos un contador de intentos
print("______________________________________________________")
print("¡¡Bienvenida al juego de AHORCADO 🕹️!!\n"
    "Si la letra introducida forma parte de la palabra secreta, podrás seguir jugando 🙌.\n"
    "Si, por el contrario, no es correcta 😞, se dibujará el ahorcado y se te restará un intento.\n"
    "Con 9 intentos fallidos el ahorcado se habrá completado y habrás perdido el juego 😭.\n"
    "🍀!!Mucha suerte!!🍀 ")

# Empieza el bucle para adivinar la palabra
while intentos < 9: # 9 son los intentos que se emplean para dibujar el ahorcado
    letra = input("Introduce una letra: ").lower() #usamos lower, para coger siempre las letras en minúsculas

    if letra in lista_palabra_secreta:
        # Si la letra es correcta, reemplaza los '*' en la posición correcta
        for i in range(len(lista_palabra_secreta)):
            if lista_palabra_secreta[i] == letra: #de i, que es nuestra palabra convertida en lista, está la letra del input
                lista_tablero[i] = letra #aqui le decimos que entonces, la sustituya en el tablero

        print(f"Letra correcta!      ➡️  {' '.join(lista_tablero)}") # el método join es para unir elementos de dos listas
                                                 # se usa el espacio, para separar los *
        
        if ''.join(lista_tablero) == palabra_secreta:
            print("🎉🎉🎉¡HAS ACERTADO LA PALABRA! 🎉🎉🎉")
            break  # Sale del bucle si se adivinó la palabra

    else:
        intentos +=1
        if intentos == 1:
            print(f"La {letra} no está en la palabra secreta, te quedan 8 intentos")
            print("       ")
            print("       ")
            print("       ")
            print("       ")
            print("       ")
            print("=======")    
        if intentos ==2:
            print(f"La {letra} no está en la palabra secreta, te quedan 7 intentos")
            print("       ")
            print("       ")
            print("       ")
            print("       ")
            print("|    ")
            print("=======")
        if intentos ==3:
            print(f"La {letra} no está en la palabra secreta, te quedan 6 intentos")
            print("       ")
            print("       ")
            print("|    ")
            print("|    ")
            print("|    ")
            print("=======")
        if intentos ==4:
            print(f"La {letra} no está en la palabra secreta, te quedan 5 intentos")
            print("       ")
            print("|    ")
            print("|    ")
            print("|    ")
            print("|    ")
            print("=======")
        if intentos ==5:
            print(f"La {letra} no está en la palabra secreta, te quedan 4 intentos")
            print("|    ")
            print("|    ")
            print("|    ")
            print("|    ")
            print("|    ")
            print("=======")
        if intentos ==6:
            print(f"La {letra} no está en la palabra secreta, te quedan 3 intentos")
            print(" *****")
            print("|    ")
            print("|    ")
            print("|    ")
            print("|    ")
            print("|    ")
            print("=======")
        if intentos ==7:
            print(f"La {letra} no está en la palabra secreta, te quedan 2 intentos")
            print(" *****")
            print("|  😒 ")
            print("|    ")
            print("|    ")
            print("|    ")
            print("|    ")
            print("=======")
        if intentos ==8:
            print(f"La {letra} no está en la palabra secreta, te queda 1 intento")
            print(" *****")
            print("|  😒 ")
            print("| /|\\ ")
            print("|    ")
            print("|    ")
            print("|    ")
            print("=======")
        if intentos ==9:
            print(f"La {letra} no está en la palabra secreta, HAS PERDIDO, la palabra era {palabra_secreta}")
            print(" *****")
            print("| 😵 ")
            print("| /|\\ ")
            print("| / \\  ")
            print("|    ")
            print("|    ")
            print("=======")
            break
        

   
```

## Juego piedra papel tijera

```python
import random

print("¡BIENVENIDO!🎉🎊¿Preparado para jugar uno de los juegos más clásicos de la historia?")
print("Te vamos a dar 10 oportunidades para elegir entre piedra, papel o tijera, vas a competir VS la computadora") 
print("Solo tienes que lograr 3 puntos para ganar....¿te atreves a jugar?🤔")
print("¡Vamos a por ello! 🎊")

def obtener_eleccion_usuario():
    opciones = ["piedra", "papel", "tijera"]
    while True:
        eleccion = input("Elige piedra, papel o tijera: ").lower()
        if eleccion in opciones:
            return eleccion
        else:
            print("Entrada no válida. Intenta de nuevo.")

def obtener_eleccion_computadora():
    return random.choice(["piedra", "papel", "tijera"])

def determinar_ganador(jugador, computadora):
    if jugador == computadora:
        return "Empate"
    elif (jugador == "piedra" and computadora == "tijera") or \
         (jugador == "tijera" and computadora == "papel") or \
         (jugador == "papel" and computadora == "piedra"):
        return "Jugador"
    else:
        return "Computadora"

def jugar():
    puntos_objetivo = 3  # Se necesitan 3 puntos para ganar
    max_rondas = 10  # Máximo de 10 rondas
    puntaje_jugador = 0
    puntaje_computadora = 0
    rondas_jugadas = 0

    while puntaje_jugador < puntos_objetivo and puntaje_computadora < puntos_objetivo and rondas_jugadas < max_rondas:
        jugador = obtener_eleccion_usuario()
        computadora = obtener_eleccion_computadora()
        print(f"Computadora eligió: {computadora}")
        print(f"Jugador eligió: {jugador}")

        ganador = determinar_ganador(jugador, computadora)
        if ganador == "Jugador":
            puntaje_jugador += 1
            print("¡Ganaste esta ronda!")
        elif ganador == "Computadora":
            puntaje_computadora += 1
            print("La computadora gana esta ronda.")
        else:
            print("Es un empate.")

        rondas_jugadas += 1
        print(f"Puntuación -> Jugador: {puntaje_jugador} | Computadora: {puntaje_computadora}")
        print(f"Rondas jugadas: {rondas_jugadas}/{max_rondas}\n")

    # Determinar el resultado final
    if puntaje_jugador > puntaje_computadora:
        print("🎉 ¡Felicidades! Ganaste el juego.")
    elif puntaje_jugador < puntaje_computadora:
        print("😢 La computadora ganó el juego.")
    else:
        print("🤝 Es un empate en el juego.")

# Iniciar el juego
jugar()
```

## Juego preguntas y respuestas

```python
class JuegoPreguntas:

    print("¡BIENVENIDO!🎉🎊.¿Preparado para poner a prueba tus conocimientos?")
    print("te vamos a hacer una serie de preguntas y tienes que elegir entre las 3 opciones que te damos....¿suena facil?🤔")
    print("¡Vamos a comprobarlo! solo te puedes equivocar 3 veces😨")
    def __init__(self):
        self.preguntas = [
            {"pregunta": "¿Cuál es la capital de Brasil?", "opciones": ["a) Madrid", "b) Berlín", "c) Brasilia"], "respuesta": "c"},
            {"pregunta": "¿Cuál es el país más pequeño del mundo?", "opciones": ["a) Brasil", "b) Vaticano", "c) Japón"], "respuesta": "b"},
            {"pregunta": "¿Cuál es el océano más grande del mundo?", "opciones": ["a) Atlántico", "b) Índico", "c) Pacífico"], "respuesta": "c"},
            {"pregunta": "¿Cuál es el país más grande del mundo?", "opciones": ["a) Rusia", "b) China", "c) Brasil"], "respuesta": "a"},
            {"pregunta": "¿Cuál es el río más largo del mundo?", "opciones": ["a) Amazonas", "b) Nilo", "c) Tajo"], "respuesta": "b"}
        ]
        self.puntuacion=0
        self.respuestas_incorrectas=0
  
    def iniciar_juego(self):  
        
        for i, p in enumerate(self.preguntas):
            print(f"\nPregunta {i + 1}: {p['pregunta']}")
            for opcion in p['opciones']:
                print(opcion)
            
            respuesta_usuario = input("Elige la respuesta (a, b, c): ").strip().lower()

            if respuesta_usuario == p['respuesta']:
                self.puntuacion += 1
                print(f"✅ ¡Correcto!🤩su puntuacion es {self.puntuacion}.La respuesta correcta era:{p['respuesta']})")
            
            else:
                self.respuestas_incorrectas+=1
                print(f"❌ Incorrecto😭. La respuesta correcta era: {p['respuesta']})")

            if  self.respuestas_incorrectas >=3:
                print("lo siento, has alcanzado el limite de preguntas incorrectas, su partida ha terminado")
                break
            print(f"🔹llevas {self.puntuacion} aciertos👍 y {self.respuestas_incorrectas} fallos👎")    

        print(f"\n🎉 Juego terminado. Tu puntuación final es: {self.puntuacion}/{len(self.preguntas)}")
juego=JuegoPreguntas()
juego.iniciar_juego()
```
