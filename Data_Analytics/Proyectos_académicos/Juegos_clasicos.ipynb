## Juego piedra papel tijera

```python
import random

print("Â¡BIENVENIDO!ğŸ‰ğŸŠÂ¿Preparado para jugar uno de los juegos mÃ¡s clÃ¡sicos de la historia?")
print("Te vamos a dar 10 oportunidades para elegir entre piedra, papel o tijera, vas a competir VS la computadora") 
print("Solo tienes que lograr 3 puntos para ganar....Â¿te atreves a jugar?ğŸ¤”")
print("Â¡Vamos a por ello! ğŸŠ")

def obtener_eleccion_usuario():
    opciones = ["piedra", "papel", "tijera"]
    while True:
        eleccion = input("Elige piedra, papel o tijera: ").lower()
        if eleccion in opciones:
            return eleccion
        else:
            print("Entrada no vÃ¡lida. Intenta de nuevo.")

def obtener_eleccion_computadora():
    return random.choice(["piedra", "papel", "tijera"])

def determinar_ganador(jugador, computadora):
    if jugador == computadora:
        return "Empate"
    elif (jugador == "piedra" and computadora == "tijera") or \
         (jugador == "tijera" and computadora == "papel") or \
         (jugador == "papel" and computadora == "piedra"):
        return "Jugador"
    else:
        return "Computadora"

def jugar():
    puntos_objetivo = 3  # Se necesitan 3 puntos para ganar
    max_rondas = 10  # MÃ¡ximo de 10 rondas
    puntaje_jugador = 0
    puntaje_computadora = 0
    rondas_jugadas = 0

    while puntaje_jugador < puntos_objetivo and puntaje_computadora < puntos_objetivo and rondas_jugadas < max_rondas:
        jugador = obtener_eleccion_usuario()
        computadora = obtener_eleccion_computadora()
        print(f"Computadora eligiÃ³: {computadora}")
        print(f"Jugador eligiÃ³: {jugador}")

        ganador = determinar_ganador(jugador, computadora)
        if ganador == "Jugador":
            puntaje_jugador += 1
            print("Â¡Ganaste esta ronda!")
        elif ganador == "Computadora":
            puntaje_computadora += 1
            print("La computadora gana esta ronda.")
        else:
            print("Es un empate.")

        rondas_jugadas += 1
        print(f"PuntuaciÃ³n -> Jugador: {puntaje_jugador} | Computadora: {puntaje_computadora}")
        print(f"Rondas jugadas: {rondas_jugadas}/{max_rondas}\n")

    # Determinar el resultado final
    if puntaje_jugador > puntaje_computadora:
        print("ğŸ‰ Â¡Felicidades! Ganaste el juego.")
    elif puntaje_jugador < puntaje_computadora:
        print("ğŸ˜¢ La computadora ganÃ³ el juego.")
    else:
        print("ğŸ¤ Es un empate en el juego.")

# Iniciar el juego
jugar()
```

## Juego preguntas y respuestas

```python
class JuegoPreguntas:

    print("Â¡BIENVENIDO!ğŸ‰ğŸŠ.Â¿Preparado para poner a prueba tus conocimientos?")
    print("te vamos a hacer una serie de preguntas y tienes que elegir entre las 3 opciones que te damos....Â¿suena facil?ğŸ¤”")
    print("Â¡Vamos a comprobarlo! solo te puedes equivocar 3 vecesğŸ˜¨")
    def __init__(self):
        self.preguntas = [
            {"pregunta": "Â¿CuÃ¡l es la capital de Brasil?", "opciones": ["a) Madrid", "b) BerlÃ­n", "c) Brasilia"], "respuesta": "c"},
            {"pregunta": "Â¿CuÃ¡l es el paÃ­s mÃ¡s pequeÃ±o del mundo?", "opciones": ["a) Brasil", "b) Vaticano", "c) JapÃ³n"], "respuesta": "b"},
            {"pregunta": "Â¿CuÃ¡l es el ocÃ©ano mÃ¡s grande del mundo?", "opciones": ["a) AtlÃ¡ntico", "b) Ãndico", "c) PacÃ­fico"], "respuesta": "c"},
            {"pregunta": "Â¿CuÃ¡l es el paÃ­s mÃ¡s grande del mundo?", "opciones": ["a) Rusia", "b) China", "c) Brasil"], "respuesta": "a"},
            {"pregunta": "Â¿CuÃ¡l es el rÃ­o mÃ¡s largo del mundo?", "opciones": ["a) Amazonas", "b) Nilo", "c) Tajo"], "respuesta": "b"}
        ]
        self.puntuacion=0
        self.respuestas_incorrectas=0
  
    def iniciar_juego(self):  
        
        for i, p in enumerate(self.preguntas):
            print(f"\nPregunta {i + 1}: {p['pregunta']}")
            for opcion in p['opciones']:
                print(opcion)
            
            respuesta_usuario = input("Elige la respuesta (a, b, c): ").strip().lower()

            if respuesta_usuario == p['respuesta']:
                self.puntuacion += 1
                print(f"âœ… Â¡Correcto!ğŸ¤©su puntuacion es {self.puntuacion}.La respuesta correcta era:{p['respuesta']})")
            
            else:
                self.respuestas_incorrectas+=1
                print(f"âŒ IncorrectoğŸ˜­. La respuesta correcta era: {p['respuesta']})")

            if  self.respuestas_incorrectas >=3:
                print("lo siento, has alcanzado el limite de preguntas incorrectas, su partida ha terminado")
                break
            print(f"ğŸ”¹llevas {self.puntuacion} aciertosğŸ‘ y {self.respuestas_incorrectas} fallosğŸ‘")    

        print(f"\nğŸ‰ Juego terminado. Tu puntuaciÃ³n final es: {self.puntuacion}/{len(self.preguntas)}")
juego=JuegoPreguntas()
juego.iniciar_juego()
```
